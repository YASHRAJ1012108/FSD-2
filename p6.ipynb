# Regression Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
dataset=pd.read_csv("auto-mpg.csv")
sns.set(font_scale=2) #incease font size
sns.regplot(x="weight",y="model year",data=dataset,color="red",marker="*",scatter_kws={'s':50})#line=reg line and dot near means good co-relation and otherwise error
plt.show()
```

```python
dataset.corr() #give relation table
```

# heatmaps

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
# sns.heatmap(data,vmin=30,vmax=70) vmin 30 less and vmax 70 max dont consider
sns.heatmap(data,cmap="tab20",annot=True)#annot give values in box and cmap gives diff colors
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
sns.heatmap(data,cmap="Pastel1",annot=True,linewidths=2,linecolor="red",cbar=False,xticklabels=False,yticklabels=False)#linewidth-borderline width of all cell,cbar-bar close
plt.show()
```

```python

dataset=pd.read_csv("Police_Department_Incidents_-_Previous_Year__2016_.csv")
print(dataset.shape)
df=dataset.iloc[0:100]
print(df.shape)
latt=37.7749
long=-122.4194
sanfras_map=folium.Map(location=[latt,long],zoom_start=12)
sanfras_map
inci=folium.map.FeatureGroup()#<folium.map.FeatureGroup at 0x142804ffd90>
for lat,long,labels in zip(df.Y,df.X,df.Category):
    inci.add_child(folium.CircleMarker([lat,long],radious=5,color="red",fill=True,fill_color="Black",fill_opacity=0.6),
                   folium.Marker([lat,long],popup=labels).add_to(sanfras_map))
sanfras_map.add_child(inci)
```

```python
state_unemp=pd.read_csv("US_Unemployment_Oct2012.csv")
state_geo="us-states.json"
usa_state=folium.Map(loaction=[48,-102],zoom_start=3)
folium.Choropleth(geo_data=state_geo,name="choropleth",data=state_unemp,columns=["State","Unemployment"],key_on="feature.id",
                 fill_color="YlGnBu",fill_opacity=0.7,line_opacity=0.2,legend_name="Unemployment Scale").add_to(usa_state)
usa_state
```
