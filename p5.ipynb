```python
#simple area chart
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,alpha=0.5,color="skyblue")
plt.plot(x,y,alpha=0.8,color="r")#used for border on area

plt.show()
```

```python
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color="green")
plt.plot(time,expense,color="red")
plt.fill_between(time,income,expense,where=(income>expense),color="green",alpha=0.25,label="Positive",interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color="red",alpha=0.25,label="Negative",interpolate=True)
plt.xlabel("time")
plt.ylabel("income/expense comparision")
plt.legend()
plt.show()

```

# scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



x1=5*np.random.rand(40)
x2=5*np.random.rand(40)+25
x3=25*np.random.rand(20)
print("x3 data=",x3,"x2 data=",x2,"x2 data=",x2)
x=np.concatenate((x1,x2,x3))

y1=5*np.random.rand(40)
y2=5*np.random.rand(40)+25
y3=25*np.random.rand(20)
y=np.concatenate((y1,y2,y3))

color_array=["b"]*40+["g"]*40+["r"]*20
plt.scatter(x,y,s=[100],marker="^",c=color_array)
plt.show()
```

```python
#pip install pywaffle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from pywaffle import Waffle
data={"phone":["Xiaomi","Samsung","Apple","Nokia","Realme"],"stock":[45,40,40,10,50]}
df=pd.DataFrame(data)
fig=plt.figure(FigureClass=Waffle,rows=15,values=df.stock,labels=list(df.phone))
```

```python
#word cloud if number repeat many times then size of that increase
#alice in wonderland txt
#pip install wordcloud==1.8.1
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS

alice_novel=open("alice_in_wonderland.txt")

alice_novel=alice_novel.read()
stopwords=set(STOPWORDS) #delete some words like he she which is repeated
stopwords.add("said")
# print(stopwords)
print(len(stopwords))#192
alice_wc=WordCloud(background_color="black",max_words=2000,stopwords=stopwords)
alice_wc.generate(alice_novel) #<wordcloud.wordcloud.WordCloud at 0x142db34bd30>
plt.figure(figsize=[20,20])
plt.imshow(alice_wc)
plt.axis("on")
plt.show()
```
